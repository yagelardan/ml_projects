```python
from sklearn.datasets import fetch_olivetti_faces
```

```python
dataset = fetch_olivetti_faces(shuffle=True)
X = dataset.data
y = dataset.target
```

```python
n_samples, n_features = X.shape
print(n_samples, n_features)
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl

def plot_face(face):
    fig = plt.figure()
    face = face.reshape(64,64)
    plt.imshow(face)
    plt.show()
print("5 first faces:")
for i in range(0,5):
    print("index: ",i)
    plot_face(X[i])
```

```python
#Recuce dimensionals using pca
from sklearn.decomposition import PCA 
pca = PCA(n_components=0.99)
X_pca = pca.fit_transform(X) #for one digit
X_pca.shape
```

```python
print("Inverse dimensional reduced image using pca:")
plot_face(pca.inverse_transform(X_pca[0]))
```
