```python
import numpy as np
import os

def sort_by_target(mnist):
    reorder_train = np.array(sorted([(target, i) for i, target in enumerate(mnist.target[:60000])]))[:, 1]
    reorder_test = np.array(sorted([(target, i) for i, target in enumerate(mnist.target[60000:])]))[:, 1]
    mnist.data[:60000] = mnist.data[reorder_train]
    mnist.target[:60000] = mnist.target[reorder_train]
    mnist.data[60000:] = mnist.data[reorder_test + 60000]
    mnist.target[60000:] = mnist.target[reorder_test + 60000]
    
    
try:
    from sklearn.datasets import fetch_openml
    mnist = fetch_openml('mnist_784', version=1, cache=True)
    mnist.target = mnist.target.astype(np.int8) # fetch_openml() returns targets as strings
    sort_by_target(mnist) # fetch_openml() returns an unsorted dataset
except ImportError:
    from sklearn.datasets import fetch_mldata
    mnist = fetch_mldata('MNIST original')
mnist["data"], mnist["target"]
```

```python
#normalize the data before PCA - https://stats.stackexchange.com/questions/69157/why-do-we-need-to-normalize-data-before-principal-component-analysis-pca
#normalized improved the accuracy by 0.39 percent
from sklearn import preprocessing
X = preprocessing.normalize(X)
```

```python
#PCA for all digits
pca_all_digits = PCA(n_components=0.90)
X_pca_all_digits = pca_all_digits.fit_transform(X) #for all digits
X_pca_all_digits.shape
```

```python
#split the data to train and test
X_pca_all_digits_train = X_pca_all_digits[:60000]#first 60000 for training
y_pca_all_digits_train = y[:60000]
X_pca_all_digits_test = X_pca_all_digits[60000:]#last 10000 for testing
y_pca_all_digits_test = y[60000:]
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
model1 = KNeighborsClassifier(n_neighbors=3)
#train the model
model1.fit(X_pca_all_digits_train,y_pca_all_digits_train)
```

```python
from sklearn import metrics
y_predict = model1.predict(X_pca_all_digits_test)
```

```python
print("Accuracy for KNN:",metrics.accuracy_score(y_pca_all_digits_test, y_predict))
```

```python
#Bagging
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bag_clf = BaggingClassifier(
    DecisionTreeClassifier(random_state=42), n_estimators=500,
    max_samples=100, bootstrap=True, n_jobs=-1, random_state=42, oob_score = True) #give oob score
bag_clf.fit(X_pca_all_digits_train, y_pca_all_digits_train)
```

```python
y_predict_bag = bag_clf.predict(X_pca_all_digits_test)
print("Accuracy for bagging:",metrics.accuracy_score(y_pca_all_digits_test, y_predict_bag)) #add another accuracy score (even though we have oob score)
```

```python
print("Accuracy of Out Of Bag data(should be close to the previous result):",bag_clf.oob_score_)
```

```python
#Apply XGBoost for MNIST
import xgboost 

param_list = [("eta", 0.08), ("max_depth", 4), ("subsample", 0.8), ("colsample_bytree", 0.8), ("objective", "multi:softmax"), ("eval_metric", "merror"), ("alpha", 8), ("lambda", 2), ("num_class", 10)]
n_rounds = 100
early_stopping = 10
```

```python
X_train = X_pca_all_digits_train[:50000]
y_train = y_pca_all_digits_train[:50000]
X_valid = X_pca_all_digits_train[50000:]
y_valid = y_pca_all_digits_train[50000:]
X_test = X_pca_all_digits_test
y_test = y_pca_all_digits_test
```

```python
d_train = xgb.DMatrix(X_train, label=y_train)
d_val = xgb.DMatrix(X_valid, label=y_valid)
eval_list = [(d_train, "train"), (d_val, "validation")]
bst = xgb.train(param_list, d_train, n_rounds, evals=eval_list, early_stopping_rounds=early_stopping, verbose_eval=True)
```

```python
d_test = xgb.DMatrix(data=X_test)
y_pred_xgb = bst.predict(d_test)
```

```python
print("Accuracy for XGBoost on test set:", metrics.accuracy_score(y_test, y_pred_xgb))
```
